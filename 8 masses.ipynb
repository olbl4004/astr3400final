```python
import numpy as np
import matplotlib.pyplot as plt
import mesa_web as m
%matplotlib inline
```

```python
low8 = '8masslowmetal.data'
high8 = '8masshighmetal.data'
```

```python
low8data = m.read_history(low8)
```

```python
high8data = m.read_history(high8)
```

```python
cgshigh8rad = np.flip(10**(high8data['log_R']) * 6.957e10)
```

```python
cgslow8rad = np.flip(10**(low8data['log_R']) * 6.957e10)
```

```python
cgshigh8mass = np.flip(high8data['star_mass'] * 1.988e33)
```

```python
cgslow8mass = np.flip(low8data['star_mass'] * 1.988e33)
```

```python
cgslow8age = np.flip(low8data['star_age'] * 3.154e7)
```

```python
cgshigh8age = np.flip(high8data['star_age'] * 3.154e7)
```

```python
plt.loglog(cgslow8age, cgslow8mass);
```

```python
plt.loglog(cgshigh8age, cgshigh8mass);
```

```python
plt.loglog(cgslow8age, cgslow8rad);
```

```python
plt.loglog(cgshigh8age, cgshigh8rad);
```

```python
plt.loglog(cgslow8age, low8data['dynamic_timescale']);
```

```python
plt.loglog(cgshigh8age, high8data['dynamic_timescale']);
```

```python
plt.loglog(cgslow8age, low8data['nuc_timescale']);
```

```python
plt.loglog(cgshigh8age, high8data['nuc_timescale']);
```

```python
plt.loglog(cgslow8age, low8data['dynamic_timescale'], c='red')
plt.loglog(cgslow8age, low8data['nuc_timescale'], c='blue')
plt.loglog(cgslow8age, low8data['kh_timescale'], c='green')
plt.xlim(1e2, 2e15)
plt.ylim(1, 2e11)
plt.show();
```

```python
plt.loglog(cgshigh8age, high8data['dynamic_timescale'], c='red')
plt.loglog(cgshigh8age, high8data['nuc_timescale'], c='blue')
plt.loglog(cgshigh8age, high8data['kh_timescale'], c='green')
plt.show();
```
