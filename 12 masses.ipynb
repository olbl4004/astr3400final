```python
import numpy as np
import matplotlib.pyplot as plt
import mesa_web as m
%matplotlib inline
```

```python
low12 = '12masslowmetal.data'
high12 = '12masshighmetal.data'
```

```python
low12data = m.read_history(low12)
```

```python
high12data = m.read_history(high12)
```

```python
cgshigh12rad = np.flip(10**(high12data['log_R']) * 6.957e10)
```

```python
cgslow12rad = np.flip(10**(low12data['log_R']) * 6.957e10)
```

```python
cgshigh12mass = np.flip(high12data['star_mass'] * 1.988e33)
```

```python
cgslow12mass = np.flip(low12data['star_mass'] * 1.988e33)
```

```python
cgslow12age = np.flip(low12data['star_age'] * 3.154e7)
```

```python
cgshigh12age = np.flip(high12data['star_age'] * 3.154e7)
```

```python
plt.loglog(cgslow12age, cgslow12mass);
```

```python
plt.loglog(cgshigh12age, cgshigh12mass);
```

```python
plt.loglog(cgslow12age, cgslow12rad);
```

```python
plt.loglog(cgshigh12age, cgshigh12rad);
```

```python
plt.loglog(cgslow12age, low12data['dynamic_timescale']);
```

```python
plt.loglog(cgshigh12age, high12data['dynamic_timescale']);
```

```python
plt.loglog(cgslow12age, low12data['nuc_timescale']);
```

```python
plt.loglog(cgshigh12age, high12data['nuc_timescale']);
```

```python
plt.loglog(cgslow12age, low12data['dynamic_timescale'], c='red')
plt.loglog(cgslow12age, low12data['nuc_timescale'], c='blue')
plt.loglog(cgslow12age, low12data['kh_timescale'], c='green')
plt.show();
```

```python
plt.loglog(cgshigh12age, high12data['dynamic_timescale'], c='red')
plt.loglog(cgshigh12age, high12data['nuc_timescale'], c='blue')
plt.loglog(cgshigh12age, high12data['kh_timescale'], c='green')
plt.show();
```
