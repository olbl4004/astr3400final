```python
import numpy as np
import matplotlib.pyplot as plt
import mesa_web as m
%matplotlib inline
```

```python
low25 = '25masslowmetal.data'
high25 = '25masshighmetal.data'
```

```python
low25data = m.read_history(low25)
```

```python
high25data = m.read_history(high25)
```

```python
cgshigh25rad = np.flip(10**(high25data['log_R']) * 6.957e10)
```

```python
cgslow25rad = np.flip(10**(low25data['log_R']) * 6.957e10)
```

```python
cgshigh25mass = np.flip(high25data['star_mass'] * 1.988e33)
```

```python
cgslow25mass = np.flip(low25data['star_mass'] * 1.988e33)
```

```python
cgslow25age = np.flip(low25data['star_age'] * 3.154e7)
```

```python
cgshigh25age = np.flip(high25data['star_age'] * 3.154e7)
```

```python
plt.loglog(cgslow25age, cgslow25mass);
```

```python
plt.loglog(cgshigh25age, cgshigh25mass);
```

```python
plt.loglog(cgslow25age, cgslow25rad);
```

```python
plt.loglog(cgshigh25age, cgshigh25rad);
```

```python
plt.loglog(cgslow25age, low25data['dynamic_timescale']);
```

```python
plt.loglog(cgshigh25age, high25data['dynamic_timescale']);
```

```python
plt.loglog(cgslow25age, low25data['nuc_timescale']);
```

```python
plt.loglog(cgshigh25age, high25data['nuc_timescale']);
```

```python
plt.loglog(cgslow25age, low25data['dynamic_timescale'], c='red')
plt.loglog(cgslow25age, low25data['nuc_timescale'], c='blue')
plt.loglog(cgslow25age, low25data['kh_timescale'], c='green')
plt.show();
```

```python
plt.loglog(cgshigh25age, high25data['dynamic_timescale'], c='red')
plt.loglog(cgshigh25age, high25data['nuc_timescale'], c='blue')
plt.loglog(cgshigh25age, high25data['kh_timescale'], c='green')
plt.show();
```
