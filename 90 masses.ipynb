```python
import numpy as np
import matplotlib.pyplot as plt
import mesa_web as m
%matplotlib inline
```

```python
low90 = '90masslowmetal.data'
high90 = '90masshighmetal.data'
```

```python
low90data = m.read_history(low90)
```

```python
high90data = m.read_history(high90)
```

```python
cgshigh90rad = np.flip(10**(high90data['log_R']) * 6.957e10)
```

```python
cgslow90rad = np.flip(10**(low90data['log_R']) * 6.957e10)
```

```python
cgshigh90mass = np.flip(high90data['star_mass'] * 1.99090e33)
```

```python
cgslow90mass = np.flip(low90data['star_mass'] * 1.99090e33)
```

```python
cgslow90age = np.flip(low90data['star_age'] * 3.154e7)
```

```python
cgshigh90age = np.flip(high90data['star_age'] * 3.154e7)
```

```python
plt.loglog(cgslow90age, cgslow90mass);
```

```python
plt.loglog(cgshigh90age, cgshigh90mass);
```

```python
plt.loglog(cgslow90age, cgslow90rad);
```

```python
plt.loglog(cgshigh90age, cgshigh90rad);
```

```python
plt.loglog(cgslow90age, low90data['dynamic_timescale']);
```

```python
plt.loglog(cgshigh90age, high90data['dynamic_timescale']);
```

```python
plt.loglog(cgslow90age, low90data['nuc_timescale']);
```

```python
plt.loglog(cgshigh90age, high90data['nuc_timescale']);
```

```python
plt.loglog(cgslow90age, low90data['dynamic_timescale'], c='red')
plt.loglog(cgslow90age, low90data['nuc_timescale'], c='blue')
plt.loglog(cgslow90age, low90data['kh_timescale'], c='green')
plt.show();
```

```python
plt.loglog(cgshigh90age, high90data['dynamic_timescale'], c='red')
plt.loglog(cgshigh90age, high90data['nuc_timescale'], c='blue')
plt.loglog(cgshigh90age, high90data['kh_timescale'], c='green')
plt.show();
```
