```python
import numpy as np
import matplotlib.pyplot as plt
import mesa_web as m
%matplotlib inline
```

```python
low50 = '50masslowmetal.data'
high50 = '50masshighmetal.data'
```

```python
low50data = m.read_history(low50)
```

```python
high50data = m.read_history(high50)
```

```python
cgshigh50rad = np.flip(10**(high50data['log_R']) * 6.957e10)
```

```python
cgslow50rad = np.flip(10**(low50data['log_R']) * 6.957e10)
```

```python
cgshigh50mass = np.flip(high50data['star_mass'] * 1.95050e33)
```

```python
cgslow50mass = np.flip(low50data['star_mass'] * 1.95050e33)
```

```python
cgslow50age = np.flip(low50data['star_age'] * 3.154e7)
```

```python
cgshigh50age = np.flip(high50data['star_age'] * 3.154e7)
```

```python
plt.loglog(cgslow50age, cgslow50mass);
```

```python
plt.loglog(cgshigh50age, cgshigh50mass);
```

```python
plt.loglog(cgslow50age, cgslow50rad);
```

```python
plt.loglog(cgshigh50age, cgshigh50rad);
```

```python
plt.loglog(cgslow50age, low50data['dynamic_timescale']);
```

```python
plt.loglog(cgshigh50age, high50data['dynamic_timescale']);
```

```python
plt.loglog(cgslow50age, low50data['nuc_timescale']);
```

```python
plt.loglog(cgshigh50age, high50data['nuc_timescale']);
```

```python
plt.loglog(cgslow50age, low50data['dynamic_timescale'], c='red')
plt.loglog(cgslow50age, low50data['nuc_timescale'], c='blue')
plt.loglog(cgslow50age, low50data['kh_timescale'], c='green')
plt.show();
```

```python
plt.loglog(cgshigh50age, high50data['dynamic_timescale'], c='red')
plt.loglog(cgshigh50age, high50data['nuc_timescale'], c='blue')
plt.loglog(cgshigh50age, high50data['kh_timescale'], c='green')
plt.show();
```
