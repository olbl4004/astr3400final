```python
import numpy as np
import matplotlib.pyplot as plt
import mesa_web as m
%matplotlib inline
```

```python
low16 = '16masslowmetal.data'
high16 = '16masshighmetal.data'
```

```python
low16data = m.read_history(low16)
```

```python
high16data = m.read_history(high16)
```

```python
cgshigh16rad = np.flip(10**(high16data['log_R']) * 6.957e10)
```

```python
cgslow16rad = np.flip(10**(low16data['log_R']) * 6.957e10)
```

```python
cgshigh16mass = np.flip(high16data['star_mass'] * 1.988e33)
```

```python
cgslow16mass = np.flip(low16data['star_mass'] * 1.988e33)
```

```python
cgslow16age = np.flip(low16data['star_age'] * 3.154e7)
```

```python
cgshigh16age = np.flip(high16data['star_age'] * 3.154e7)
```

```python
plt.loglog(cgslow16age, cgslow16mass);
```

```python
plt.loglog(cgshigh16age, cgshigh16mass);
```

```python
plt.loglog(cgslow16age, cgslow16rad);
```

```python
plt.loglog(cgshigh16age, cgshigh16rad);
```

```python
plt.loglog(cgslow16age, low16data['dynamic_timescale']);
```

```python
plt.loglog(cgshigh16age, high16data['dynamic_timescale']);
```

```python
plt.loglog(cgslow16age, low16data['nuc_timescale']);
```

```python
plt.loglog(cgshigh16age, high16data['nuc_timescale']);
```

```python
plt.loglog(cgslow16age, low16data['dynamic_timescale'], c='red')
plt.loglog(cgslow16age, low16data['nuc_timescale'], c='blue')
plt.loglog(cgslow16age, low16data['kh_timescale'], c='green')
plt.show();
```

```python
plt.loglog(cgshigh16age, high16data['dynamic_timescale'], c='red')
plt.loglog(cgshigh16age, high16data['nuc_timescale'], c='blue')
plt.loglog(cgshigh16age, high16data['kh_timescale'], c='green')
plt.show();
```
